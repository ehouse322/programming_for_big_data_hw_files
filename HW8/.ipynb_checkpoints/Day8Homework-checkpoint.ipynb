```python
def split_color_image(image_filename):
    '''
    input:
        image_filename - str, name of image file
    output:
        red_slice: numpy.ndarray, 2d array containing the red component
        green_slice - numpy.ndarray, 2d array containing the green component
        blue_slice - numpy.ndarray, 2d array containing the blue component
    '''
    image = imread(image_filename)
    red_slice = image[:, :, 0]
    green_slice = image[:, :, 1]
    blue_slice = image[:, :, 2]
    
    return red_slice, green_slice, blue_slice
```

    def average_brightness(image_filename):

        input:
            image_filename - str, name of image file

            avg - numpy.ndarray

# Question 3  Color switch

Write a function that accepts a color image and creates a new image that has all red and green values switched.

**Hint:**
- you can't just reverse the order like the class excercise

    def switch_red_and_green(image_filename):
        '''
        input:
            image_filename - str, name of image file
        output:
            recolored_image : (numpy.ndarray)
                Color image resulting from switching red and
                green values of color_image_array
        '''

```python
def switch_red_and_green(image_filename):
    '''
    input:
      image_filename - str, name of image file
    
    output:
      recolored_image : (numpy.ndarray)
          Color image resulting from switching red and
          green values of color_image_array
          
    '''
    original_image = imread(image_filename)
    recolored_image = copy.deepcopy(original_image)
    
    red_slice, green_slice, blue_slice = split_color_image(image_filename)
    
    recolored_image[:,:,0] = green_slice
    recolored_image[:,:,1] = red_slice
    
    plt.imshow(recolored_image)
    plt.show()
 
    return recolored_image
```

# Question 4: Fraction of Image

Write a function that accepts a color image and returns the fraction of the image that contains blue values larger than a given threshold. 

    def blue_fraction_of_image(image_filename, threshold=100):
        '''
        input:
        -----------
        image_filename : (numpy.ndarray)
            Color image as a numpy 3D array

        threshold : (int)
            Threshold for the blue values. Default is 100.

        Returns:
        --------
        blue_fraction : (float)
            Fraction of pixels of color_image_array that have
            a blue value larger than threshold.
        '''

```python
def blue_fraction_of_image(image_filename, threshold=100):
    '''
    input:
    -----------
    image_filename : (numpy.ndarray)
        Color image as a numpy 3D array

    threshold : (int)
        Threshold for the blue values. Default is 100.

    Returns:
    --------
    blue_fraction : (float)
        Fraction of pixels of color_image_array that have
        a blue value larger than threshold.
    '''
    image = imread(image_filename)
    triv_mask = image >= 0
    mask = image > threshold
    
    triv_blue_slice = triv_mask[:,:,2]
    blue_slice = mask[:,:,2]
    blue_total = blue_slice.sum()
    
    total = np.sum(triv_blue_slice)
    blue_fraction = blue_total / total
    
    return blue_fraction
    
```

```python
blue_fraction_of_image('../HW8/images/1895-First_Communion.-11.jpg', 100)
```

# Question 5

There is a microscope image of some cells in the `cells.png` image. Write an algorithm to count the number of cells. 

You should be aiming for a count from 60-110 cells at the low-end.

    def count_cells(image_filename):
        '''
        Parameters:
        * image_filename - file name of image
        Output:
        * cell_count - int, number of cells counted in the image
        '''

```python
def count_cells(image_filename):
    '''
    Parameters:
    * image_filename - file name of image
    Output:
    * cell_count - int, number of cells counted in the image
    '''
    image = imread(image_filename)
    mask = image > 0.1
    avg_cell_size = 50 * 40
    
    cell_pixel_total = mask.sum()
    #print(cell_pixel_total)
    cell_count = int(cell_pixel_total / avg_cell_size)
    
    return cell_count
    
    
    
```

```python
count_cells('../HW8/cells.png')
```
